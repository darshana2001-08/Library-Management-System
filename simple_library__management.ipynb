```python
# Sample library catalogue as a global variable
library_catalogue = {
    'B001': {'Title': 'The Great Gatsby', 'Author': 'F. Scott Fitzgerald', 'Category': 'Fiction', 'Renting Availability': True, 'Purchasing Price': 10.0, 'Items Available': 5},
    'B002': {'Title': '1984', 'Author': 'George Orwell', 'Category': 'Dystopian', 'Renting Availability': True, 'Purchasing Price': 15.0, 'Items Available': 3},
    'B003': {'Title': 'To Kill a Mockingbird', 'Author': 'Harper Lee', 'Category': 'Fiction', 'Renting Availability': False, 'Purchasing Price': 12.0, 'Items Available': 0},
     'B004': {'Title': 'The Art of Comp. Programming', 'Author': 'Donald Knuth', 'Category': 'Technology', 'Renting Availability': True, 'Purchasing Price': 30.0, 'Items Available': 3},
    'B005': {'Title': 'A Brief History of Time', 'Author': 'Stephen Hawking', 'Category': 'Science', 'Renting Availability': True, 'Purchasing Price': 20.0, 'Items Available': 6},
    'B006': {'Title': 'Clean Code', 'Author': 'Robert C. Martin', 'Category': 'Technology', 'Renting Availability': False, 'Purchasing Price': 25.0, 'Items Available': 7},
    'B007': {'Title': 'The Catcher in the Rye', 'Author': 'J.D. Salinger', 'Category': 'Fiction', 'Renting Availability': True, 'Purchasing Price': 14.0, 'Items Available': 3},
    'B008': {'Title': 'Cosmos', 'Author': 'Carl Sagan', 'Category': 'Science', 'Renting Availability': True, 'Purchasing Price': 18.0, 'Items Available': 5},
    'B009': {'Title': 'The Pragmatic Programmer', 'Author': 'A. Hunt, D. Thomas', 'Category': 'Technology', 'Renting Availability': True, 'Purchasing Price': 22.0, 'Items Available': 4},
    'B010': {'Title': '1984', 'Author': 'George Orwell', 'Category': 'Fiction', 'Renting Availability': True, 'Purchasing Price': 11.0, 'Items Available': 8},
    'B011': {'Title': 'Moby-Dick', 'Author': 'Herman Melville', 'Category': 'Fiction', 'Renting Availability': True, 'Purchasing Price': 15.0, 'Items Available': 4},
    'B012': {'Title': 'The Selfish Gene', 'Author': 'Richard Dawkins', 'Category': 'Science', 'Renting Availability': False, 'Purchasing Price': 17.0, 'Items Available': 2},
    'B013': {'Title': 'Guns, Germs, and Steel', 'Author': 'Jared Diamond', 'Category': 'History', 'Renting Availability': True, 'Purchasing Price': 20.0, 'Items Available': 3},
    'B014': {'Title': 'The Road', 'Author': 'Cormac McCarthy', 'Category': 'Fiction', 'Renting Availability': True, 'Purchasing Price': 14.0, 'Items Available': 5},
    'B015': {'Title': 'The Alchemist', 'Author': 'Paulo Coelho', 'Category': 'Fiction', 'Renting Availability': True, 'Purchasing Price': 12.0, 'Items Available': 7},
    'B016': {'Title': 'The Book Thief', 'Author': 'Markus Zusak', 'Category': 'Fiction', 'Renting Availability': True, 'Purchasing Price': 10.0, 'Items Available': 6},
    'B017': {'Title': 'The Hunger Games', 'Author': 'Suzanne Collins', 'Category': 'Fiction', 'Renting Availability': True, 'Purchasing Price': 16.0, 'Items Available': 4},
    'B018': {'Title': 'The Lord of the Rings', 'Author': 'J.R.R. Tolkien', 'Category': 'Fantasy', 'Renting Availability': True, 'Purchasing Price': 25.0, 'Items Available': 3},
    'B019': {'Title': 'The Da Vinci Code', 'Author': 'Dan Brown', 'Category': 'Mystery', 'Renting Availability': True, 'Purchasing Price': 18.0, 'Items Available': 5},
    'B020': {'Title': 'The Kite Runner', 'Author': 'Khaled Hosseini', 'Category': 'Historical Fiction', 'Renting Availability': True, 'Purchasing Price': 13.0, 'Items Available': 6},
    'B021': {'Title': 'The Fault in Our Stars', 'Author': 'John Green', 'Category': 'Young Adult', 'Renting Availability': True, 'Purchasing Price': 11.0, 'Items Available': 7},
    'B022': {'Title': 'The Girl with the Dragon Tattoo', 'Author': 'Stieg Larsson', 'Category': 'Mystery', 'Renting Availability': True, 'Purchasing Price': 17.0, 'Items Available': 4},
    'B023': {'Title': 'The Book Thief', 'Author': 'Markus Zusak', 'Category': 'Historical Fiction', 'Renting Availability': True, 'Purchasing Price': 12.0, 'Items Available': 5},
    'B024': {'Title': 'The Chronicles of Narnia', 'Author': 'C.S. Lewis', 'Category': 'Fantasy', 'Renting Availability': True, 'Purchasing Price': 20.0, 'Items Available': 6},
    'B025': {'Title': 'The Hobbit', 'Author': 'J.R.R. Tolkien', 'Category': 'Fantasy', 'Renting Availability': True, 'Purchasing Price': 19.0, 'Items Available': 7},
    'B026': {'Title': 'The Hunger Games', 'Author': 'Suzanne Collins', 'Category': 'Young Adult', 'Renting Availability': True, 'Purchasing Price': 15.0, 'Items Available': 5},
    'B027': {'Title': 'The Lord of the Rings', 'Author': 'J.R.R. Tolkien', 'Category': 'Fantasy', 'Renting Availability': True, 'Purchasing Price': 25.0, 'Items Available': 3},
    'B028': {'Title': 'The Da Vinci Code', 'Author': 'Dan Brown', 'Category': 'Mystery', 'Renting Availability': True, 'Purchasing Price': 18.0, 'Items Available': 5},
    'B029': {'Title': 'The Kite Runner', 'Author': 'Khaled Hosseini', 'Category': 'Historical Fiction', 'Renting Availability': True, 'Purchasing Price': 13.0, 'Items Available': 6},
    'B030': {'Title': 'The Fault in Our Stars', 'Author': 'John Green', 'Category': 'Young Adult', 'Renting Availability': True, 'Purchasing Price': 11.0, 'Items Available': 7},
    'B031': {'Title': 'The Girl with the Dragon Tattoo', 'Author': 'Stieg Larsson', 'Category': 'Mystery', 'Renting Availability': True, 'Purchasing Price': 17.0, 'Items Available': 4},
    'B032': {'Title': 'The Book Thief', 'Author': 'Markus Zusak', 'Category': 'Historical Fiction', 'Renting Availability': True, 'Purchasing Price': 12.0, 'Items Available': 5},
    'B033': {'Title': 'The Chronicles of Narnia', 'Author': 'C.S. Lewis', 'Category': 'Fantasy', 'Renting Availability': True, 'Purchasing Price': 20.0, 'Items Available': 6},
    'B034': {'Title': 'The Hobbit', 'Author': 'J.R.R. Tolkien', 'Category': 'Fantasy', 'Renting Availability': True, 'Purchasing Price': 19.0, 'Items Available': 7},
    'B035': {'Title': 'The Hunger Games', 'Author': 'Suzanne Collins', 'Category': 'Young Adult', 'Renting Availability': True, 'Purchasing Price': 15.0, 'Items Available': 5},
    'B036': {'Title': 'The Lord of the Rings', 'Author': 'J.R.R. Tolkien', 'Category': 'Fantasy', 'Renting Availability': True, 'Purchasing Price': 25.0, 'Items Available': 3},
    'B037': {'Title': 'The Da Vinci Code', 'Author': 'Dan Brown', 'Category': 'Mystery', 'Renting Availability': True, 'Purchasing Price': 18.0, 'Items Available': 5},
    'B038': {'Title': 'The Kite Runner', 'Author': 'Khaled Hosseini', 'Category': 'Historical Fiction', 'Renting Availability': True, 'Purchasing Price': 13.0, 'Items Available': 6},
    'B039': {'Title': 'The Fault in Our Stars', 'Author': 'John Green', 'Category': 'Young Adult', 'Renting Availability': True, 'Purchasing Price': 11.0, 'Items Available': 7},
    'B040': {'Title': 'The Girl with the Dragon Tattoo', 'Author': 'Stieg Larsson', 'Category': 'Mystery', 'Renting Availability': True, 'Purchasing Price': 17.0, 'Items Available': 4},
    'B041': {'Title': 'The Book Thief', 'Author': 'Markus Zusak', 'Category': 'Historical Fiction', 'Renting Availability': True, 'Purchasing Price': 12.0, 'Items Available': 5},
    'B042': {'Title': 'The Chronicles of Narnia', 'Author': 'C.S. Lewis', 'Category': 'Fantasy', 'Renting Availability': True, 'Purchasing Price': 20.0, 'Items Available': 6},
    'B043': {'Title': 'The Hobbit', 'Author': 'J.R.R. Tolkien', 'Category': 'Fantasy', 'Renting Availability': True, 'Purchasing Price': 19.0, 'Items Available': 7},
    'B044': {'Title': 'The Hunger Games', 'Author': 'Suzanne Collins', 'Category': 'Young Adult', 'Renting Availability': True, 'Purchasing Price': 15.0, 'Items Available': 5},
    'B045': {'Title': 'The Lord of the Rings', 'Author': 'J.R.R. Tolkien', 'Category': 'Fantasy', 'Renting Availability': True, 'Purchasing Price': 25.0, 'Items Available': 3},
    'B046': {'Title': 'The Da Vinci Code', 'Author': 'Dan Brown', 'Category': 'Mystery', 'Renting Availability': True, 'Purchasing Price': 18.0, 'Items Available': 5},
    'B047': {'Title': 'The Kite Runner', 'Author': 'Khaled Hosseini', 'Category': 'Historical Fiction', 'Renting Availability': True, 'Purchasing Price': 13.0, 'Items Available': 6},
    'B048': {'Title': 'The Fault in Our Stars', 'Author': 'John Green', 'Category': 'Young Adult', 'Renting Availability': True, 'Purchasing Price': 11.0, 'Items Available': 7},
    'B049': {'Title': 'The Girl with the Dragon Tattoo', 'Author': 'Stieg Larsson', 'Category': 'Mystery', 'Renting Availability': True, 'Purchasing Price': 17.0, 'Items Available': 4},
    'B050': {'Title': 'The Book Thief', 'Author': 'Markus Zusak', 'Category': 'Historical Fiction', 'Renting Availability': True, 'Purchasing Price': 12.0, 'Items Available': 5},


}


shopping_cart = {}

def display_catalogue():
    print("\nLibrary Catalogue")
    print("-" * 70)
    print(f"{'ID':<8}{'Title':<30}{'Author':<25}{'Category':<15}{'Renting':<15}{'Price ($)':<15}{'Available':<15}")
    print("-" * 70)
    for book_id, details in library_catalogue.items():
        renting_availability = "Yes" if details['Renting Availability'] else "No"
        print(f"{book_id:<8}{details['Title']:<30}{details['Author']:<25}{details['Category']:<15}{renting_availability:<15}{details['Purchasing Price']:<15}{details['Items Available']:<15}")
    print("-" * 70)

def quick_sort_catalogue(catalogue):
    if len(catalogue) <= 1:
        return catalogue

    pivot_key = catalogue[0]['Category']
    pivot = catalogue[0]

    left = [x for x in catalogue[1:] if (x['Category'] > pivot_key) or (x['Category'] == pivot_key and x['Title'] < pivot['Title'])]
    right = [x for x in catalogue[1:] if not ((x['Category'] > pivot_key) or (x['Category'] == pivot['Category'] and x['Title'] < pivot['Title']))]

    return quick_sort_catalogue(left) + [pivot] + quick_sort_catalogue(right)

def quick_sort_and_display_catalogue():
    sorted_catalogue = quick_sort_catalogue(list(library_catalogue.values()))
    print("\n*********Sorted Library Catalogue **********\n")
    print("-" * 70)
    print(f"{'Title':<30}{'Author':<25}{'Category':<15}{'Renting':<15}{'Price ($)':<15}{'Available':<15}")
    print("-" * 70)
    for details in sorted_catalogue:
        renting_availability = "Yes" if details['Renting Availability'] else "No"
        print(f"{details['Title']:<30}{details['Author']:<25}{details['Category']:<15}{renting_availability:<15}{details['Purchasing Price']:<15}{details['Items Available']:<15}")
    print("-" * 70)


def is_valid_book_id(book_id):
    if book_id in library_catalogue:
        return True
    else:
        print("Invalid Book ID. Please try again.")
        return False

def is_valid_quantity(book_id, quantity):
    available = library_catalogue[book_id]['Items Available']
    if quantity > 0 and quantity <= available:
        return True
    else:
        print("Invalid quantity. Please try again.")
        return False

def renting_period(period):
    if 30 <= period <= 90:
        return True
    else:
        print("Invalid renting period. Must be between 30 and 90 days.")
        return False

def add_book_to_cart():
    while True:
        book_id = input("Enter Book ID (or type 'END' to finish): ")
        if book_id == "END":
            return
        if not is_valid_book_id(book_id):
            continue

        action = input("Would you like to 'Buy' or 'Rent' the book? ").strip().lower()
        if action not in ['buy', 'rent']:
            print("Invalid action. Please enter 'Buy' or 'Rent'.")
            continue

        if action == 'buy':
            quantity = int(input("Enter quantity to buy: "))
            if not is_valid_quantity(book_id, quantity):
                continue

            # Add to shopping cart
            shopping_cart[book_id] = {'Action': 'Buy', 'Quantity': quantity, **library_catalogue[book_id]}
            library_catalogue[book_id]['Items Available'] -= quantity

        elif action == 'rent':
            rent_period = int(input("Enter renting period (30-90 days): "))
            if not renting_period(rent_period):
                continue

            # Add to shopping cart
            shopping_cart[book_id] = {'Action': 'Rent', 'Renting Period': rent_period, **library_catalogue[book_id]}
            library_catalogue[book_id]['Renting Availability'] = False

        print(f"{action.capitalize()} action completed for {library_catalogue[book_id]['Title']}.")

def show_cart():
    print("\nShopping Cart")
    print("-" * 70)
    print(f"{'ID':<8}{'Title':<30}{'Author':<25}{'Action':<15}{'Quantity/Rent Days':<20}{'Price ($)':<15}")
    print("-" * 70)
    total_cost = 0
    total_cost_with_gst = 0
    gst_rate = 1.10  # 18% GST

    for book_id, details in shopping_cart.items():
        if details['Action'] == 'Buy':
            cost = details['Purchasing Price'] * details['Quantity']
            total_cost += cost
            total_cost_with_gst += cost * (1 + gst_rate)  # Calculate total cost including GST
            print(f"{book_id:<8}{details['Title']:<30}{details['Author']:<25}{details['Action']:<15}{details['Quantity']:<20}{cost:<15}")
        elif details['Action'] == 'Rent':
            rent_price = (details['Purchasing Price'] * 0.3) * (details['Renting Period'] // 30)
            total_cost += rent_price
            total_cost_with_gst += rent_price * (1 + gst_rate)  # Calculate total rent price including GST
            print(f"{book_id:<8}{details['Title']:<30}{details['Author']:<25}{details['Action']:<15}{details['Renting Period']:<20}{rent_price:<15}")

    print("-" * 70)
    print(f"{'Total Cost (without GST):':<70}{total_cost:<15}")
    print(f"{'Total Cost (with GST):':<70}{total_cost_with_gst:<15}")
    print("-" * 70)


def main_menu():
    while True:
        print("\nMain Menu")
        print("1. Display Catalogue")
        print("2. Sort Catalogue")
        print("3. Add Book to Cart")
        print("4. Show Shopping Cart")
        print("5. Exit")

        choice = input("Select an option (1-5): ")

        if choice == '1':
            display_catalogue()
        elif choice == '2':
            quick_sort_and_display_catalogue()
        elif choice == '3':
            add_book_to_cart()
        elif choice == '4':
            show_cart()
        elif choice == '5':
            print("Exiting the system. Goodbye!")
            break
        else:
            print("Invalid selection. Please try again.")

# Start the library management system
main_menu()
```
